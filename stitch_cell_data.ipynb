```python
import pandas as pd
from pathlib import Path
from matplotlib import pyplot as plt
import numpy as np
```

```python
sample = 'IPICRC088T1_8plex'

tile_pos_path = Path("/home/austin/data/mibi/deepcell/ark-analysis/data/",sample,"tile_metadata.txt")
single_cell_output_path = Path("/home/austin/data/mibi/deepcell/ark-analysis/data/",sample,"single_cell_output//cell_table_arcsinh_transformed.csv")
```

```python
tile_pos_data = pd.read_csv(tile_pos_path,header=None)
tile_pos_data = tile_pos_data.rename(columns={0:"fov",1:"x1",2:"x2",3:"y1",4:"y2"})
```

```python
cell_data = pd.read_csv(single_cell_output_path)
```

```python
f = plt.figure(figsize=(20,20))

for t in tile_pos_data.iterrows():
   
    xvals = cell_data[cell_data.fov == t[1].fov]['centroid-0'].values
    yvals = cell_data[cell_data.fov == t[1].fov]['centroid-1'].values
    
    xvals += t[1].x1
    yvals += t[1].y1
    
    cell_data.loc[cell_data.fov == t[1].fov,'centroid-0'] = xvals
    cell_data.loc[cell_data.fov == t[1].fov,'centroid-1'] = yvals
    
    plt.scatter(xvals,yvals)
    plt.text(np.mean(xvals), np.mean(yvals), t[1].fov)
plt.show()
```

```python
savedir = Path("/media/austin/IPI_8plex_project/8plex/analysis/cell_data_tables", sample)

try:
    savedir.mkdir()
except FileExistsError:
    pass

cell_data.to_csv("/media/austin/IPI_8plex_project/8plex/analysis/cell_data_tables/"+sample+"/cell_table_arcsinh_transformed_stitched.csv")
```
